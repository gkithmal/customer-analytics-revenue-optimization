```python
import numpy as np

# Creating arrays (lists of numbers)
ages = np.array([25, 30, 22, 45, 33])
print("Customer ages:", ages)

# Quick math on ALL numbers at once
print("Average age:", ages.mean())
print("Ages + 1 year:", ages + 1)

# Generate random numbers (we'll use this for fake data)
random_prices = np.random.randint(10, 100, size=5 )
print("Random product prices:", random_prices)
```

```python
import pandas as pd

# Create a simple customer table
customer_data = {
    'name': ['Alice', 'Bob', 'Charlie'],
    'age': [25, 30, 22],
    'spent': [150, 220, 80]
}

df = pd.DataFrame(customer_data)
print("Customer Table:")
print(df)

# Quick analysis
print("\nAverage spending:", df['spent'].mean())
print("Customers over 25:", df[df['age'] > 25])
```

```python
# Let's start with simple data you can understand
# Data
ages = [25, 30, 22, 45, 33, 28, 35, 40, 29, 31]
spending = [150, 220, 80, 400, 180, 120, 250, 350, 100, 200]
print("Ages:", ages)
print("Spending:", spending)
```

```python
import matplotlib.pyplot as plt

# Scatter plot shows relationships
plt.figure(figsize=(8, 5))
plt.scatter(ages, spending, color='red', alpha=0.6)
plt.title('Age vs Spending')
plt.xlabel('Age')
plt.ylabel('Amount Spent ($)')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Simple bar chart
plt.figure(figsize=(8, 5))  # Set size of chart
plt.bar(range(len(ages)), ages, color='skyblue')
plt.title('Customer Ages')
plt.xlabel('Customer Number')
plt.ylabel('Age')
plt.show()
```

```python
import matplotlib.pyplot as plt

names = ['Alice', 'Bob', 'Charlie']
ages = [24, 30, 18]

plt.figure(figsize=(6, 4))  # Set chart size
plt.bar(range(len(names)), ages, color='orange')  # Create bars

# Set x-axis labels to names
plt.xticks(range(len(names)), names)

# Add labels and title
plt.xlabel("Person")
plt.ylabel("Age")
plt.title("Age of People")

plt.show()
```

```python
# Histogram shows distribution
plt.figure(figsize=(8, 5))
plt.hist(ages, bins=5, color='lightgreen', alpha=0.7)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Customers')
plt.show()
```
